```python
# Load in first year
raw_data = loadarff('2year.arff')
df_data = pd.DataFrame(raw_data[0])

# Columns are not included with data, taken and formatted from source
df_data.columns = columns
df_data["bankruptcy"] = [int(b) for b in df_data['bankruptcy']]
df_data
```

```python
# Exploration time!

is_bankrupt = df_data['bankruptcy'] == 1
bankrupt_banks = df_data[is_bankrupt]
print(f'Shape of df: {bankrupt_banks.shape}')
print(f'Bankrupt percentage: {len(bankrupt_banks) / len(df_data)}')
bankrupt_banks.head()
```
